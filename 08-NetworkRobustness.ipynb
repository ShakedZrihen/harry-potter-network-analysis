```python
import networkx as nx
%matplotlib inline 
import matplotlib.pyplot as plt
```

```python
G = nx.read_edgelist('edges_list_pos_weight.txt', delimiter=',', data = [('weight', int)]) 
```

```python
plt.figure(figsize=(7,7))
nx.draw_networkx(G, alpha=0.7, with_labels=True, edge_color='.4')
```

```python
print('nx.is_connected(G) ->', nx.is_connected(G))
print('nx.node_connectivity(G) ->', nx.node_connectivity(G))
print('nx.minimum_node_cut(G) ->', nx.minimum_node_cut(G)) 
```

```python
print('nx.edge_connectivity(G) ->', nx.edge_connectivity(G))
print('nx.minimum_edge_cut(G) ->', nx.minimum_edge_cut(G)) 
```
