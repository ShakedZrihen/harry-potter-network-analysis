```python
import networkx as nx
%matplotlib inline 
import matplotlib.pyplot as plt
import operator 

# full friends network
G = nx.read_gpickle('HarryPotterFullNetwork-friends.gpickle')
```

```python
### link prediction ###
## based on common neigbors ##

print('common_neighbors = [(e[0], e[1], len(list(nx.common_neighbors(G, e[0], e[1])))\n for e in nx.non_edges(G)]\nsorted(common_neighbors, key=operator.itemgetter(2), reverse=True) ->')
 
common_neighbors = [(e[0], e[1], len(list(nx.common_neighbors(G, e[0], e[1]))))
 for e in nx.non_edges(G)]
sorted(common_neighbors, key=operator.itemgetter(2), reverse=True)
```

```python
### Jaccard coefficient ###

print('preds = nx.jaccard_coefficient(G)\njaccard_coefficient = [(u, v, p) for u, v, p in preds]\nsorted(jaccard_coefficient, key=operator.itemgetter(2), reverse=True) ->')
preds = nx.jaccard_coefficient(G)
jaccard_coefficient = [(u, v, p) for u, v, p in preds]
sorted(jaccard_coefficient, key=operator.itemgetter(2), reverse=True)
```

```python
### Resource allocation ###

print('L = list(nx.resource_allocation_index(G))\nsorted(L, key=operator.itemgetter(2), reverse=True) ->')

L = list(nx.resource_allocation_index(G)) 
sorted(L, key=operator.itemgetter(2), reverse=True)
```

```python
### Preferential attachment ###

print('L = list(nx.preferential_attachment(G))\nsorted(L, key=operator.itemgetter(2), reverse=True) ->')

L = list(nx.preferential_attachment(G)) 
sorted(L, key=operator.itemgetter(2), reverse=True)
```
