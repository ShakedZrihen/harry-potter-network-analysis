```python
import networkx as nx
%matplotlib inline 
```

```python
# Adjacency List - from txt

G = nx.read_adjlist('adjacency_list.txt', delimiter=',')
nx.draw_networkx(G)
```

```python
# to Adjacency Matrix

G2 = nx.to_numpy_matrix(G)
print(G2)
```

```python
# edge view from adjacency matrix

G = nx.Graph(G2)
G.edges() 
```

```python
# edges list - from txt

G = nx.read_edgelist('edges_list.txt', delimiter=',', data = [('weight', int)]) 
G.edges(data=True) 
```

```python
# to matrix

G2 = nx.to_numpy_matrix(G)
print(G2)
```

```python
# create data frame from csv

import pandas as pd

G_df = pd.read_csv('cleanRelations.csv', delimiter=',')
G_df.head()
```

```python
# create network from data frame

G = nx.from_pandas_edgelist(G_df, 'name1', 'name2', edge_attr='relation')
nx.draw_networkx(G)
```
