```python
import networkx as nx
%matplotlib inline 
import pandas as pd
import matplotlib.pyplot as plt

# full network
G_df = pd.read_csv('friends_relations.csv')
G = nx.from_pandas_edgelist(G_df, 'name1', 'name2')
```

```python
### link prediction ###
## based on common neigbors ##

c_m = [(e[0], e[1], len(list(nx.common_neighbors(G, e[0], e[1]))))
 for e in nx.non_edges(G)]
```

```python
import operator 
sorted(c_m, key=operator.itemgetter(2), reverse=True)
```

```python
### Jaccard coefficient ###

preds = nx.jaccard_coefficient(G)
j_c = [(u, v, p) for u, v, p in preds]
sorted(j_c, key=operator.itemgetter(2), reverse=True)
```

```python
### Resource allocation ###

L = list(nx.resource_allocation_index(G)) 
sorted(L, key=operator.itemgetter(2), reverse=True)
```
