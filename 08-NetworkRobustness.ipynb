```python
import networkx as nx
%matplotlib inline 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
G = nx.read_edgelist('edges_list_pos_weight.txt', delimiter=',', data = [('weight', int)]) 
```

```python
plt.figure(figsize=(7,7))
nx.draw_networkx(G, alpha=0.7, with_labels=True, edge_color='.4')
```

```python
print('is_connected: ', nx.is_connected(G))
print('node connectivity: ', nx.node_connectivity(G))
print('minimum node cut :', nx.minimum_node_cut(G)) 
```

```python
print('edge connectivity: ', nx.edge_connectivity(G))
print('minimum edge cut: ', nx.minimum_edge_cut(G)) 
```

```python
print('edge connectivity: ', nx.edge_connectivity(G))
print('minimum edge cut: ', nx.minimum_edge_cut(G)) 
```
