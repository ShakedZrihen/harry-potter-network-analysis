```python
import networkx as nx
%matplotlib inline 
import matplotlib.pyplot as plt
```

```python
G = nx.read_edgelist('edges_list_pos_weight.txt', delimiter=',', data = [('weight', int)]) 
plt.figure(figsize=(7,7))
nx.draw_networkx(G, alpha=0.7, with_labels=True, edge_color='.4')
```

```python
# is connected?

nx.is_connected(G) 
```

```python
# disconnect network

G.remove_edges_from([('Harry Potter', 'Ron Weasley'), ('Harry Potter', 'Hermione Granger')]) 

nx.draw_networkx(G, alpha=0.7, with_labels=True, edge_color='.4')
nx.is_connected(G)
```

```python
# connected conponent

print(nx.number_connected_components(G))
print(sorted(nx.connected_components(G))) 
```

```python
print(nx.node_connected_component(G, 'Harry Potter')) 
```
