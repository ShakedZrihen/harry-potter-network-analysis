```python
import networkx as nx
%matplotlib inline 
import matplotlib.pyplot as plt
import operator
```

```python
G = nx.read_edgelist('edges_list_pos_weight.txt', delimiter=',', data = [('weight', int)], create_using=nx.DiGraph) 
plt.figure(figsize=(7,7))
nx.draw_networkx(G, nx.circular_layout(G), alpha=0.7, with_labels=True, edge_color='.4')
```

```python
### degree centrality ###
print('nx.degree_centrality(G) ->\n', nx.degree_centrality(G))
print('\nnx.in_degree_centrality(G) ->\n', nx.in_degree_centrality(G))
print('\nnx.out_degree_centrality(G) ->\n', nx.out_degree_centrality(G))
```

```python
# G2 is not directed

G2 = nx.read_edgelist('edges_list_pos_weight.txt', delimiter=',', data = [('weight', int)], create_using=nx.Graph) #not directed
plt.figure(figsize=(7,7))
nx.draw_networkx(G2, nx.circular_layout(G2), alpha=0.7, with_labels=True, edge_color='.4')
```

```python
### closeness centrality ###
## closeness centrality(node) = 1 / sum(distance from node to all other nodes - (neo4j.com) ##

print('nx.closeness_centrality(G2) ->')
nx.closeness_centrality(G2) 
```

```python
### nodes betweenness centrality ###

print('nx.betweenness_centrality(G2, normalized=True) ->')
nx.betweenness_centrality(G2, normalized=True) 
```

```python
### edges betweenness centrality ###

print('nx.edge_betweenness_centrality(G2, normalized=True) ->')
nx.edge_betweenness_centrality(G2) 
```

```python
### PageRank ###

print('nx.pagerank(G) ->')
nx.pagerank(G) 
```

```python
############# Full Network ###############

G1 = nx.read_gpickle('HarryPotterFullNetwork-friends.gpickle')
plt.figure(figsize=(15,15))
nx.draw_networkx(G1, nx.spring_layout(G1), alpha=0.7, with_labels=True, edge_color='.4')
```

```python
### degree centrality ###
nx.degree_centrality(G1)
```

```python
node_size = [G1.degree(v) * 20 for v in G1] # Size will be the degree
plt.figure(3,figsize=(15,15))

## Color nodes in house colors
house_colors = {'Slytherin':'darkgreen', 'Ravenclaw':'darkblue',
 'Hufflepuff':'goldenrod','Gryffindor': 'darkred', 'other':'lightgray'}
node_color = [house_colors[character['house']]
 if character['house'] in house_colors else
  house_colors['other'] for name, character in G1.nodes(data=True) ]

## draw graph
nx.draw_networkx(G1, nx.circular_layout(G1),
 font_size=10, width=0.5, edge_color='lightgray',
  node_size=node_size, node_color=node_color)
```

```python
### closeness centrality ###
## closeness centrality(node) = 1 / sum(distance from node to all other nodes - (neo4j.com) ##
nx.closeness_centrality(G1) 
```

```python
### nodes betweenness centrality ###
nx.betweenness_centrality(G1, normalized=True) 
```

```python
### edges betweenness centrality ###
nx.edge_betweenness_centrality(G1) 
```

```python
### PageRank ###
nx.pagerank(G1)
```

```python
node_size = [nx.pagerank(G1)[v] * 20000 for v in G1] # Size will be the pagerank
plt.figure(3,figsize=(15,15))

## Color nodes in house colors
house_colors = {'Slytherin':'darkgreen', 'Ravenclaw':'darkblue',
 'Hufflepuff':'goldenrod','Gryffindor': 'darkred', 'other':'lightgray'}
node_color = [house_colors[character['house']]
 if character['house'] in house_colors else
  house_colors['other'] for name, character in G1.nodes(data=True) ]

## draw graph
nx.draw_networkx(G1, nx.circular_layout(G1),
 font_size=10, width=0.5, edge_color='lightgray',
  node_size=node_size, node_color=node_color)
```

```python
################# exercise - 9.2 #################

import collections
degree_sequence = sorted([d for n, d in G1.degree()], reverse=True)
degreeCount = collections.Counter(degree_sequence)
deg, cnt = zip(*degreeCount.items())
plt.bar(deg, cnt, width=0.80, color=(0.2, 0.4, 0.6, 0.6))
plt.title("Degree Histogram")
plt.ylabel("Count")
plt.xlabel("Degree")
ax = plt.subplot()
ax.set_xticks([d + 0.4 for d in deg])
ax.set_xticklabels(deg)
plt.show() 
```

```python
G3 = nx.barabasi_albert_graph(1000000, 1) 
degree_sequence = sorted([d for n, d in G3.degree()], reverse=True)
degreeCount = collections.Counter(degree_sequence)
deg, cnt = zip(*degreeCount.items())
plt.bar(deg, cnt, width=0.80, color=(0.2, 0.4, 0.6, 0.6))
plt.title("Degree Histogram")
plt.ylabel("Count")
plt.xlabel("Degree")
plt.xscale('log')
plt.yscale('log')
plt.show() 
```
