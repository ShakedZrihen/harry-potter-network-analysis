```python
# show 
from networkx.algorithms import bipartite 

B = nx.Graph()
B.add_nodes_from(['Harry Potter', 'Ron Weasley', 'Draco Malfoy', 'Neville Longbottom'], bipartite = 0)
B.add_nodes_from(['Dark Arts', 'Defence Against the Dark Arts', 'Herbology', 'Flying'], bipartite = 1)

B.add_edges_from([('Dark Arts', 'Draco Malfoy'), ('Defence Against the Dark Arts', 'Harry Potter'), ('Defence Against the Dark Arts', 'Neville Longbottom'), ('Herbology', 'Neville Longbottom'), ('Herbology', 'Ron Weasley'), ('Flying', 'Harry Potter'), ('Flying', 'Ron Weasley'), ('Flying', 'Draco Malfoy'), ('Flying', 'Neville Longbottom')])

students, classes = bipartite.sets(B)
pos = dict()
pos.update( (n, (0, i)) for i, n in enumerate(students) ) # put nodes from students at x=0
pos.update( (n, (1, i)) for i, n in enumerate(classes) ) # put nodes from Y at classes=1

nx.draw_networkx(B, pos=pos, font_size=10, alpha=0.7, node_color='yellow')
```

```python
# check if bipartite

bipartite.is_bipartite(B) 
```

```python
# change and check again

B.add_edge('Harry Potter', 'Ron Weasley')
print(bipartite.is_bipartite(B)) 

B.remove_edge('Harry Potter', 'Ron Weasley')
print(bipartite.is_bipartite(B)) 
```

```python
X1 = set(['Harry Potter', 'Ron Weasley', 'Draco Malfoy', 'Neville Longbottom'])
X2 = set(['Dark Arts', 'Defence Against the Dark Arts', 'Herbology', 'Flying'])
X3 = set(['Dark Arts', 'Defence Against the Dark Arts', 'Herbology', 'Flying', 'Harry Potter'])

print(bipartite.is_bipartite_node_set(B, X1))
print(bipartite.is_bipartite_node_set(B, X2))
print(bipartite.is_bipartite_node_set(B, X3)) 
```
