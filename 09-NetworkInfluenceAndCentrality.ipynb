```python
import networkx as nx
%matplotlib inline 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
G = nx.read_edgelist('edges_list_pos_weight.txt', delimiter=',', data = [('weight', int)], create_using=nx.DiGraph) 
plt.figure(figsize=(7,7))
nx.draw_networkx(G, alpha=0.7, with_labels=True, edge_color='.4')
```

```python
### degree centrality ###

print('degree centrality:', nx.degree_centrality(G))
print('in degree centrality:', nx.in_degree_centrality(G))
print('out degree centrality:', nx.out_degree_centrality(G))
```

```python
### closeness centrality ###
## closeness centrality(node) = 1 / sum(distance from node to all other nodes - (neo4j.com) ##
nx.closeness_centrality(G) 
```

```python
### nodes betweenness centrality ###
nx.betweenness_centrality(G, normalized=True) 
```

```python
### edges betweenness centrality ###
nx.edge_betweenness_centrality(G) 
```

```python
### PageRank ###
nx.pagerank(G) 
```
